```python
#Tic Tac Toe

from IPython.display import clear_output

def display_board(board):
    clear_output()
    print(board[1:4])
    print(board[4:7])
    print(board[7:10])

def player_input():
    #initial choice
    choice = "Wrong"
    
    #make sure choice is X O
    while choice not in ['X','O']:
        choice = input(f'Player 1: Do you want to be X or O? ')
        
    if choice == 'X':
        return('X','O')
    else:
        return('O','X')

def place_marker(board, marker, position):
    board[position] = marker

def win_check(board, mark):
    win = [(1,2,3),(4,5,6),(7,8,9),(1,4,7),(2,5,8),(3,6,9),(1,5,9),(3,5,7)]
    for w in win:
        if board[w[0]] == board[w[1]] and board[w[1]] == board[w[2]] and board[w[0]] == mark:
            return True
    return False

import random

def choose_first():
    if random.randint(0, 1) == 0:
        return 'Player 2'
    else:
        return 'Player 1'

def space_check(board, position):
    return board[position] == ''

def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
            return False
    return True

def player_choice(board):
    position = 0
    
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board, position):
        position = int(input('Choose your next position: (1-9) '))
        
    return position

def replay():
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
```

```python
while True:
    #print board
    board = ['']*10
    board[0] = '#'
    display_board(board)

    #decide player
    player1, player2 = player_input()
    turn = choose_first()
    print(turn + ' will go first.')
    
    #ready to play
    play_game = input('Are you ready to play? Enter Yes or No.')
    
    if play_game.lower()[0] == 'y':
        game_on = True
    else:
        game_on = False
    
    #place input on board
    while (game_on):
        if turn == 'Player 1':
            display_board(board)
            place_marker(board, player1, player_choice(board))
            
            if win_check(board, player1):
                display_board(board)
                print('You win!')
                game_on=False
            else:
                if full_board_check(board):
                    display_board(board)
                    print('Game is tie!')
                    break
                else:
                    turn = 'Player 2'
        else:
            display_board(board)
            place_marker(board, player2, player_choice(board))
            
            if win_check(board, player2):
                display_board(board)
                print('Player 2 wins!')
                game_on=False
            else:
                if full_board_check(board):
                    display_board(board)
                    print('Game is tie!')
                    break
                else:
                    turn = 'Player 1'
    
    #ask to replay
    if not replay():
        print('Thank you for playing! Quiting..')
        break
```
